```python
import numpy as np
from numpy import ndarray

students: ndarray = np.array([[78, 92, 68, 95, 91],
                              [65, 75, 63, 84, 86]])

students[0]#取得第1位學生資料
students[1]#取得第2位學生資料
students[1,:3]#取得第2位學生資料，前3科分數
students[0,-3:]#取得第1位學生資料，後3科分數

```

```python
subjects:list[str]=['chinest','english','math','history','discover']
subjects
```

```python
import matplotlib.pyplot as plt

figure = plt.figure()
axes=figure.add_subplot()
axes.plot(subjects, students[0],label="stu1")
axes.plot(subjects, students[1],label="stu2")
axes.legend()
axes.set_title("stu1 vs stu 2")
axes.set_xlabel("subject")
axes.set_ylabel("scores")
plt.show()
plt.close()
```
